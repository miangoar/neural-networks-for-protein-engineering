```python
!pip install biopython --upgrade
from Bio import SeqIO
import pylab
```

```python
# Load raw files from supplementary materials of Fantini et al.
g1 = [len(i) for i in SeqIO.parse("bla.GEN1_ccs.alllength.pep", "fasta")]
g5 = [len(i) for i in SeqIO.parse("bla.GEN5_ccs.alllength.pep", "fasta")]
g12 = [len(i) for i in SeqIO.parse("bla.GEN12_ccs.longest.orf.pep", "fasta")]
```

Plot the length of the sequences as histograms

```python
# G1 
pylab.hist(g1, bins=40)
pylab.title("Generation 1 | Total sequences = %i  \n Lengths %i to %i" % (len(g1), min(g1), max(g1)))
pylab.xlabel("Sequence length (aa)")
pylab.ylabel("Count")
pylab.show()
```

```python
pylab.hist(g5, bins=40)
pylab.title("Generation 5 | Total sequences = %i \n Lengths %i to %i" % (len(g5), min(g5), max(g5)))
pylab.xlabel("Sequence length (aa)")
pylab.ylabel("Count")
pylab.show()
```

```python
pylab.hist(g12, bins=40)
pylab.title(" Generation 12 | Total sequences = %i \n Lengths %i to %i" % (len(g12), min(g12), max(g12)))
pylab.xlabel("Sequence length (aa)")
pylab.ylabel("Count")
pylab.show()
```
