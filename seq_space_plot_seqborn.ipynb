```python
import pandas as pd 
import seaborn as sns
sns.set(rc={'figure.figsize':(25,10)})
sns.set_style('white')
sns.set_context("paper", font_scale = 2)
```

```python
tsne_data =  pd.read_csv("all_tsne_proyections.csv")
tsne_data = tsne_data.drop([0])
```

```python
sns.jointplot(
    data=tsne_data,
    x="component_0", 
    y="component_1", 
    hue="label",
    height = 10,
    kind="hist",
    alpha=0.5
)
```
